```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("product_sales_dataset_final.csv")
```

```python
df.head()
```

```python
df = df.drop(columns=[
        "Order_ID", "Order_Date", "Customer_Name",
        "City", "State", "Region", "Country", "Product_Name"
    ]
)
```

```python
df.head()
```

```python
df.columns
```

```python
df.isnull().sum()
```

```python
df["comp_price_subcat"] = (
    df.groupby("Sub_Category")[" Unit_Price "]
      .transform("median")
)
```

```python
df["comp_price_cat"] = (
    df.groupby("Category")[" Unit_Price "]
      .transform("median")
)
```

```python
global_median_price = df[" Unit_Price "].median()
```

```python
df["Competitor_Price"] = (
    df["comp_price_subcat"]
      .fillna(df["comp_price_cat"])
      .fillna(global_median_price)
)
```

```python
df.head()
```

```python
df["elasticity_index"] = df["Competitor_Price"] / df[" Unit_Price "]
```

```python
df.head()
```

```python
df["elasticity_index"].describe()
```

```python
df["elasticity_index"] = df["elasticity_index"].clip(0.1,3)
```

```python
df["elasticity_index"].describe()
```

```python
df.to_csv("cleaned_product_sales.csv",index=False)
```
